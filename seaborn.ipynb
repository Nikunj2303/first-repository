```python
import  seaborn as sns
import pandas as pd
fmri= pd.read_csv("fmri.csv")
print(fmri.head())
sns.lineplot(x = "timepoint", y="signal", hue="region",
            style="event", data=fmri)
```

```python
iris = pd.read_csv("iris.csv")

sns.displot(iris['sepal_length'],color='k')

```

```python
df = pd.read_csv("anscombe.csv")
sns.lmplot(x="x",y="y",data = df)

sns.swarmplot(x ="petal_length", y="species",data=iris) 

sns.stripplot(x="species", y="petal_length",data=iris)
```

```python
tip= pd.read_csv("tips.csv")
sns.barplot(x='sex', y='total_bill',data=tip)

sns.countplot(x='sex', data=tip)
sns.boxplot(x='day', y='total_bill',data=tip,hue='smoker')
```
